We define some dictionnaries and sets to have the correct data that is present in both the Results file and the Output file in our dataset.

```python
gene_id_to_symbol = dict(list(RawResults[['gene_id','gene_symbol']].values))
```

```python
snp_rsid_to_position = dict(list(RawResults[['ld_snp_rsID','GRCh38_pos']].values))
```

```python
rsID_gene_tuple_results = set((x[0],x[1]) for x in list(RawResults[['ld_snp_rsID','gene_id']].values))
```

```python
rsID_gene_tuple_output = set((x[0],x[1]) for x in list(RawOutput[['SNP_ID','Gene_ID']].values))
```

```python
difference = list(rsID_gene_tuple_output - rsID_gene_tuple_results)
```

```python
old_ids = []
for entry in difference:
    old_ids.append(RawOutput.loc[RawOutput['SNP_ID'] == entry[0]].loc[RawOutput['Gene_ID'] == entry[1]].index.values) 
```

```python
list_old_ids = []
for ids in old_ids:
    for id in ids:
        list_old_ids.append(id)
```

```python
Output = RawOutput.drop(list_old_ids)
```

We create new columns for the gene names in the international standards, the positions in the genome, the p-values for the eQTL study and the p-values for the GWAS study.

```python
Output['Genes'] = [ gene_id_to_symbol[gene] for gene in Output['Gene_ID']]
```

```python
Output['Positions'] = [ snp_rsid_to_position[snp_rsid] for snp_rsid in Output['SNP_ID']]
```

```python
eQTL_p_value = [ 1 - OptResults.loc[(data[6],data[2]),'GTEx_'+data[4]].values[0] for data in Output.values]
```

```python
p_values = dict(RawResults.loc[RawResults['ls_snp_is_gwas_snp'] == 1,['ld_snp_rsID','gwas_pvalue']].values)
```

```python
Output['eQTL_p_value'] = eQTL_p_value
```

```python
Output['GWAS_p_value'] = [p_values[snp_rsid] for snp_rsid in Output['SNP_ID']]
```

```python
temp = Output.replace({'eQTL_p_value': {0: inf}})
temp2 = temp.replace({'GWAS_p_value': {0: inf}})
Output = temp2.copy()
```

We now have our final dataset that we can use to generate the figures we need.

# Manhattan plots

Let's now dive into the Manhattan plots. We create new columns to have -log10(p-value) and we group our dataset by gene and by tissue. Our goal is to have a Manhattan plot for a gene expression in a specific tissue.

```python
Output['-log10(SNP CLPP)']=[-log10(data) for data in Output['CLPP_at_that_SNP'].values]
```

```python
Output['-log10(eQTL p-value)']=[-log10(data) for data in Output['eQTL_p_value'].values]
```

```python
Output['-log10(GWAS_p_value)']=[-log10(data) for data in Output['GWAS_p_value'].values]
```

```python
GroupedByTissue = [ Output.loc[Output['Tissue'] == str(tissue) ] for tissue in Unique_Tissues ]
```

```python
GroupedByGeneTissue = [ group.loc[group['Genes'] == str(gene) ] for gene in Unique_Genes for group in GroupedByTissue ]
```

As we're using a toy dataset, we want to find the Manhattan plots with the most points, because the vast majority of Manhattan plots has 1 point in our dataset.

```python
Lengths = []
for i in range(len(GroupedByGeneTissue)):
    Lengths.append(len(GroupedByGeneTissue[i]))
```

```python
len(GroupedByGeneTissue)
```

```python
df = pd.DataFrame(Lengths)
df.max()
```

```python
a = df.loc[df[0] == 7]
a
```

We have 68332 possible Manhattan plots, the maximum number of point sin a single one is 7 and we choose a random subset <ith 7 points to generate our Manhattan Plots.

```python
data = GroupedByGeneTissue[68301]
```

```python
data
```

```python
colors = ['red','green','blue', 'yellow','purple','black','cyan']
plt.scatter('Positions', '-log10(SNP CLPP)', data=data)
plt.xlabel('Positions')
plt.ylabel('-log10(SNP CLPP)')
plt.title('Expression of ' + str(data['Genes'].values[0]) + ' in ' + str(data['Tissue'].values[0]))
fig = plt.gcf()
fig.set_size_inches(16.5, 6.5)
plt.show()
```

```python
colors = ['red','green','blue', 'yellow','purple','black','cyan']
plt.scatter('Positions', '-log10(eQTL p-value)', data=data)
plt.xlabel('Positions')
plt.ylabel('-log10(eQTL p-value)')
plt.title('Expression of ' + str(data['Genes'].values[0]) + ' in ' + str(data['Tissue'].values[0]))
fig = plt.gcf()
fig.set_size_inches(16.5, 6.5)
plt.show()
```

```python
colors = ['red','green','blue', 'yellow','purple','black','cyan']
plt.scatter('Positions', '-log10(GWAS_p_value)', data=data)
plt.xlabel('Positions')
plt.ylabel('-log10(GWAS_p_value)')
plt.title('Expression of ' + str(data['Genes'].values[0]) + ' in ' + str(data['Tissue'].values[0]))
fig = plt.gcf()
fig.set_size_inches(16.5, 6.5)
plt.show()
```

We see that for our different Manhattan plots, we have different values for each study we use. We can see that the p-values for the eQTL study is quite high, because -log10(pvalue) is quite low. For the GWAS study, we have good results, with values near 5, meaning we have p-values as low as 10^-5. For our toy dataset, this is quite a good performance.
